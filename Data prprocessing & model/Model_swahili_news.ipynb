```
import pandas as pd
import numpy as np
np.random.seed(2021)
```

```
df = pd.read_csv('train_pro.csv')
test_df = pd.read_csv('test_pro.csv')
df.head()
```

```
df.isnull().sum()
```

```
test_df.isnull().sum()
```

```
df = df.fillna("Habari za kitaifa")
test_df = test_df.fillna("Habari za kitaifa")
```

```
df.isnull().sum()
```

```
test_df.isnull().sum()
```

```
df.head()
```

```
test_df.head()
```

```
df.shape, test_df.shape
```

```
df = pd.get_dummies(df, prefix_sep = "_", columns = ['category'])
```

```
df.head(2)
```

```
df['labels'] = list(zip(df.category_kitaifa.tolist(), df.category_michezo.tolist(), df.category_burudani.tolist(), df.category_uchumi.tolist(),  df.category_kimataifa.tolist(), df.category_afya.tolist()))
df['text'] = df['processed'].apply(lambda x: x.replace('\n', ' '))

df.head()
```

```
from sklearn.model_selection import train_test_split

train_df, eval_df = train_test_split(df, test_size=0.2)
```

```
from simpletransformers.classification import MultiLabelClassificationModel

model = MultiLabelClassificationModel('roberta', 'roberta-large', num_labels=6, args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 5, 'max_seq_length': 512})
```

```
#model = MultiLabelClassificationModel('roberta', 'roberta-large', num_labels=6)
```

```
# Train the model
model.train_model(train_df)
```

```
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```

```
to_predict = test_df.processed.apply(lambda x: x.replace('\n', ' ')).tolist()
preds, outputs = model.predict(to_predict)

sub_df = pd.DataFrame(outputs, columns=['kitaifa', 'michezo', 'burudani','uchumi', 'kimataifa', 'afya'])


sub_df['id'] = test_df['id']
sub_df = sub_df[['id', 'kitaifa', 'michezo', 'burudani','uchumi', 'kimataifa', 'afya']]

sub_df.to_csv('outputs/submissionwinhereyougo.csv', index=False)
```
